```python
from IPython import get_ipython
from IPython.display import display
# %%
# Import libraries
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import urllib.request  # Import urllib.request instead of just urllib

# Use urllib.request.urlretrieve to download the image
urllib.request.urlretrieve("https://static.wikia.nocookie.net/sonys-marvel-universe/images/9/9d/Sydney_Sweeney.jpg/revision/latest/scale-to-width-down/1000?cb=20240213114708", "local-filename.jpg")

# Reading an Image
image = Image.open('local-filename.jpg')
# Convert the image to grayscale using the convert() method
image_gray = image.convert('L') # Use image.convert('L') to convert to grayscale

# Define a threshold value (e.g., 128)
threshold = 128

# Convert the grayscale image to binary using the point() method
image_binary = image_gray.point(lambda p: p > threshold and 255)

plt.imshow(image) # Display the grayscale image using imshow with cmap='gray'
plt.show() # Show the plot

plt.imshow(image_gray, cmap='gray') # Display the grayscale image using imshow with cmap='gray'
plt.show() # Show the plot

plt.imshow(image_binary, cmap='gray') # Display the binary image using imshow with cmap='gray'
plt.show() # Show the plot
```
